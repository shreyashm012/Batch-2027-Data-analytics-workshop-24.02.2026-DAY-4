```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris


iris = load_iris()

# Convert to DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add target column
df['target'] = iris.target

# Map numeric target to species names
df['species'] = df['target'].map({
    0: 'Setosa',
    1: 'Versicolor',
    2: 'Virginica'
})

# Show first 5 rows
print("First 5 Rows of Dataset:")
print(df.head())


class_counts = df['species'].value_counts()

plt.figure()
class_counts.plot(kind='bar')
plt.title("Count of Each Iris Class")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()


avg_values = df.groupby('species')['sepal length (cm)'].mean()

plt.figure()
avg_values.plot(kind='bar')
plt.title("Average Sepal Length per Class")
plt.xlabel("Species")
plt.ylabel("Average Sepal Length (cm)")
plt.show()


plt.figure()
df['sepal length (cm)'].plot(kind='hist', bins=10)
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()


df.hist(figsize=(8, 6))
plt.show()
```
